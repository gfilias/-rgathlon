```python
graph = {
    '1': [('2', 7), ('3', 7),('4', 7)],
    '2': [('5', 7), ('6', 7)],
    '3': [('6', 5)],
    '4': [('6', 2)],
    '5': [('7', 3)],
    '6': [('7', 6)],
    '7': [('8', 0),('9', 0)],
    '8': [('10', 1)],
    '9': [('10', 1)],
    '10': [('11', 1),('12', 1)],
    '11': [('13', 1)],
    '12': [('13', 1)],
    '13': [('14', 1),('15', 1)],
    '14': [('16', 5)],
    '15': [('16', 3)],
    '16': [('17', 0)],
    '17': [('18', 10),('19', 10)],
    '18': [('20', 10)],
    '19': [('20', 16)],
    '20': [('21', 0),('22', 0),('23', 0)],
    '21': [('24', 3),('25', 3)],
    '22': [('24', 3),('25', 3)],
    '23': [('26', 13)],
    '24': [('26', 3)],
    '25': [('26', 10)],
    '26': [('27', 0),('28', 0)],
    '27': [('29', 5)],
    '28': [('30', 18)],
    '29': [('30', 11)],
    '30': [('31', 0),('32', 0)],
    '31': [('33', 12)],
    '32': [('33', 12)],
    '33': []    
}

def max_edge_sum(graph, start, end, memo={}):
    # Returns the maximum sum of all edges from start to end
    if (start, end) in memo:
        return memo[(start, end)]
    if start == end:
        return 0
    if start not in graph:
        return float('-inf')
    max_sum = float('-inf')
    for node, weight in graph[start]:
        max_sum = max(max_sum, max_edge_sum(graph, node, end, memo) + weight)
    memo[(start, end)] = max_sum
    return max_sum

def find_paths(graph, start, end, path=[], paths=[]):
    # Returns all paths from start to end
    path = path + [start]
    if start == end:
        paths.append(path)
    if start not in graph:
        return []
    for node, _ in graph[start]:
        find_paths(graph, node, end, path, paths)
    return paths

def max_edge_path(graph, start, end):
    # Returns the path with the maximum sum of all edges from start to end
    max_sum = max_edge_sum(graph, start, end)
    all_paths = find_paths(graph, start, end)
    max_paths = []
    for p in all_paths:
        if len(p) > 0:
            path_sum = 0
            for i in range(len(p) - 1):
                for node, weight in graph[p[i]]:
                    if node == p[i + 1]:
                        path_sum += weight
                        break
            if path_sum == max_sum:
                max_paths.append(p)
    return max_paths

# Example usage
max_paths = max_edge_path(graph, '1', '33')
max_sum = max_edge_sum(graph, '1', '33')
print("Max sum:", max_sum)
print("Paths with max sum:", max_paths)
```
